```python
from matplotlib.font_manager import FontProperties
fontP = FontProperties()
fontP.set_size('medium')
params = {'mathtext.default': 'regular' }  
def get_mean_mae(env, prefix_e):
    prefix_s = 1
    res = []
    for i in range(prefix_s, prefix_e):
        path = 'checkpoints/'+env+'/timestamp_prediction/prefix_'+str(i)
        with open(path + '/results2.txt', "r") as fout:
            res.append(float(fout.read()[:8]))
    return res

envs = ['helpdesk','bpi_12_w','traffic_ss']
anchor = ['-','-.','--']
prefix_e_lis = [14,  73, 13]
i = 0
for env, prefix_e in zip(envs, prefix_e_lis):
    print(env)
    res = get_mean_mae(env, prefix_e)
    res = np.array(res)
    if env=='traffic_ss':
        res = res/100
    plt.plot(res,anchor[i])
    i +=1 
envs = ['Helpdesk','BPI12W','$Traffic_{ss}$']
plt.legend(envs)


plt.ylabel("MAE(days)")   
plt.xlabel("Sequence length")
```
